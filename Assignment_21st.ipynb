Writing a file and inserting 5 lines using a loop

```python
with open("loop.txt","w") as f:
  count =0
  while count <5 :
    f.write("first line has Python\n")
    f.write("second line\n")
    f.write("third line has Python\n")
    f.write("fourth line has Python\n")
    f.write("fifth line\n")
    count +=1
```

```python
with open("loop.txt","r") as f:
  print(f.read())
```

Read lines consisting only "Python"

```python
with open("loop.txt","r") as f:
    for lines in f:
      if "Python" in lines:
        print(lines)
```

Count characters in file

```python
with open("count.txt", "w") as f:
  f.write("I have 2800 characters")
```

```python
with open("count.txt","r") as f:
  char=f.read()
  print("Total characters:", len(char))
```

Append current date, time in a *file*

```python
from datetime import datetime
with open("log.txt","a") as log:
  now = datetime.now()
  log.write("\nDate & Time:"+now.strftime("%d/%m/%Y %I:%M:%S %p"))
```

```python
with open("log.txt","r") as log:
  print(log.read())
```

Creating folder and saving 3 files in it


```python
import os


Assignments = "Assignments"

if not os.path.exists(Assignments):
    os.mkdir(Assignments)

print(f"Folder '{Assignments}' created successfully.")
```

```python
import shutil

files=["count.txt","log.txt","loop.txt"]
for file in files:
  shutil.move(file, os.path.join(Assignments, file))
```

```python
for file in os.listdir(Assignments):
  if file.endswith (".txt"):
    print(file)
```
